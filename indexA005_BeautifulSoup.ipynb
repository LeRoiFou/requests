Analyser le contenu d'une page avec beautifulsoup - 00:24:30

L'instanciation de la librairie Beautifulsoup comprend deux arguments :
- le contenu du texte de l'URL effectué à partir de la librairie requests
- le parser (analyse du fichier)

Il existe 3 parser :
- 'lxml-xml' : analyse des fichiers de type .xml
- 'html.parser' : analyse des fichiers .html écrit en langage C plus rapide que le parser suivant mais fonctionne moins bien si les balises du fichier html sont mal formatés
- 'html5lib' : analyse des fichiers .html écrit en langage Python donc plus lent que le précédent parser 'html.parser', mais plus tolérant dans les erreurs de balisage

Date : 16-02-24

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://books.toscrape.com'
```

```python
# Récupération des données de l'url
response = requests.get(url)
```

```python
# Instanciation du contenu du texte de l'URL ci-avant 
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# AFfichage de l'HTML indenté avec prettify()
print(soup.prettify())
```
