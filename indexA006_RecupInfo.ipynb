```python
import requests
from bs4 import BeautifulSoup
import pprint
```

```python
url = 'https://books.toscrape.com'
```

```python
# Récupération des données de l'url
response = requests.get(url)
```

```python
# Instanciation du contenu du texte de l'URL ci-avant 
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Récupération du 1er body uniquement avec la fonction find() en str
body = soup.find('body')
print(body.prettify()) # indentation du résultat avec prettify()
```

```python
# Récupération de tous les body avec find_all() dans une liste
body = soup.find_all('body')
print(body)
```

```python
# Récupération de la 1ère image avec find()
images = soup.find('img')
pprint.pprint(images.prettify()) # indentation du résultat avec prettify()
```

```python
# Récupération de toutes les images avec find_all()
images = soup.find_all('img')
pprint.pprint(images)
```

[Récupération du contenu à gauche de la page @ avec pour titre "Livres" (voir ci-dessous)](https://books.toscrape.com/index.html)

```python
# Contenu html du div 'aside'
aside = soup.find('aside')
for child in aside.children:
    print(child.name)
# Les retours 'None' sont dans le fichier html des retours à la ligne...
```

```python
# Contenu html du div 'aside' sans les 'None'
aside = soup.find('aside')
for child in aside.children:
    if child.name:
        print(child.name)
```

```python
# Afficher toutes les balises 'a'
links = soup.find_all('a')
pprint.pprint(links)
```
