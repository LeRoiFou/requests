```python
# Assignation du sélecteur css avec la classe "title" 
# permettant de récupérer tous les titres des jeux
game_list = find_all(S('.title'))
```

```python
game_list
```

```python
# Compréhension de liste permettant de récupérer uniquement 
# le titre de chaque jeu
games = [item.web_element.text for item in game_list]
games
```
