Récupérer le contenu d'une page avec requests - 00:18:30

Date : 16-02-24

```python
import requests
import pprint
```

```python
# Récupérer les informations d'une page
response = requests.get('https://www.google.com')
```

```python
# Code retour de la requête
print(response.status_code)
```

```python
# Contenu de la page
pprint.pprint(response.text)
```

```python
# Récupération du contenu du texte dans un fichier html
with open('results/index.html', 'w') as f:
    f.write(response.text)
```

```python
# Levées d'erreurs :
url = 'http://urlnexistepasjelaiinvente.com'
try:
    response = requests.get(url)
    response.raise_for_status()
except requests.exceptions.HTTPError as errh :
    print(f"HTTP Error : {errh}") # erreur sur le protocole HTTP
except requests.exceptions.ConnectionError as errc :
    print(f"Error connecting : {errc}") # erreur sur le protocole HTTP
except requests.exceptions.Timeout as errt :
    print(f"Timeout Error : {errt}") # trop de temps de chargement
except requests.exceptions.RequestException as err:
    print(f"OOps : Something Else {err}") # va savoir !
```
